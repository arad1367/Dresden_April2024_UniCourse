### 6. Matplotlib

```python
# Simple plot with matplotlib
x = np.linspace(1, 100, 200)
y = np.linspace(101, 200, 200)

# Make our first plot
plt.plot(x, y, linewidth=2, markersize=12, c="blue")
plt.show();
```

```python
# OOP to make a plot

# data --> x, y

fig, ax = plt.subplots(figsize=(5, 5))

# Make plot with OOP
ax.plot(x**3-1, y**2+3, color='green', marker='o', linestyle='dashed')

# Customize our plot
ax.set(title="Example Plot Curve",
       xlabel="x",
       ylabel="y")

# Save the figure
fig.savefig("saved_fig.png")
```

### How to make a time series plot

```python
time_plot = pd.DataFrame(data=np.random.normal(size=(1000, 5)),
                         index = pd.date_range("01/06/2022", periods=1000),
                         columns=["BTC", "ETH", "SOL", "ADA", "DOGE"])
time_plot = time_plot.cumsum()
plt.figure()
time_plot.plot();
```
